```python
import pandas as pd
from langdetect import detect
```

```python
def read_book_data():
    dfs = []
    for i in range(45):
        dfs.append(pd.read_csv(f'Get Data/books_data_{i}.csv'))
    book_data = pd.concat(dfs,ignore_index=True)
    return book_data
```

```python
book_data = read_book_data()
```

```python
book_data.info()
```

```python
book_data.duplicated().sum()
```

```python
book_data['genres'] = book_data['genres'].apply(lambda x : x.replace(
    '[',''
).replace(']','').replace("'",''))
```

```python
def is_english(text):
    try:
        return detect(str(text)) == 'en'
    except:
        return False
```

```python
book_data = book_data[book_data['summary_clean'].apply(is_english)]
```

```python
book_data['analysis_row'] = book_data['name'] +' '+ book_data['genres'] + ' ' + book_data['summary_clean']
```
